# **Попытка искать по ключам и выводить строки после ключа**

```python
import re
import pdfplumber

def find_text_after_keyword(pdf_path, keywords):
    texts = []

    with pdfplumber.open(pdf_path) as pdf:
        for page_number, page in enumerate(pdf.pages, 1):
            text = page.extract_text()

            for keyword in keywords:
                # Используем регулярное выражение для поиска целого абзаца текста после ключевого слова
                pattern = re.compile(rf"{re.escape(keyword)}\s*([\s\S]*?)(?=\n\n|\n$|$)", re.IGNORECASE)
                matches = re.findall(pattern, text)

                for match in matches:
                    texts.append({
                        'text': match.strip(),
                        'page': page_number
                    })

    return texts

pdf_path = "/content/drive/MyDrive/Хакатон Екатеринбург/Expert/Обучающий размеченый/¦а¦-¦¬¦+¦¦¦¬ ¦Я¦Ф тДЦ1 ¦з¦-TБTВTМ тДЦ2 ¦Ш¦¬¦-.1.00001-21_¦Х¦У¦н-26404.pdf"

keywords = ['объекту', 'объект', 'наименование']
texts = find_text_after_keyword(pdf_path, keywords)

for text in texts:
    print("абзац текста после ключевого слова:")
    print(text['text'])
    print("Страница:", text['page'])
    print()
```

# **Автоматическая проверка на орфографию, исправление текста с учетом орфографии**

В этом примере добавлен новый шаг spell_check_lines, который использует pyaspeller для проверки орфографии строк. Затем подчеркнутые строки создаются на основе откорректированных строк.

Обратите внимание, что для использования pyaspeller вам может понадобиться установить его с помощью pip install pyaspeller и иметь доступ к Интернету для работы с сервисом проверки орфографии Яндекса.

```python
import os
import fitz
import difflib
from pyaspeller import YandexSpeller

# Путь к исходному PDF-файлу
pdf_path = "/content/drive/MyDrive/Хакатон Екатеринбург/Expert/новый пдф/clean.pdf"

# Путь к новому файлу
output_directory = "/content/drive/MyDrive/Хакатон Екатеринбург/Expert/новый пдф с исправленным текстом "
os.makedirs(output_directory, exist_ok=True)
output_path = os.path.join(output_directory, "marked.pdf")

def find_similar_lines(text, header_text, threshold=0.6):
    lines = []
    header_text = header_text.lower()

    for line in text:
        line_lower = line.lower()  # Изменение регистра строки
        similarity = difflib.SequenceMatcher(None, header_text, line_lower).ratio()

        if similarity > threshold:
            lines.append(line)

    return lines

def spell_check_lines(lines):
    speller = YandexSpeller()
    corrected_lines = []

    for line in lines:
        check_result = speller.spelled(line)
        if not check_result:
            corrected_lines.append(line)

    return corrected_lines

def create_marked_pdf(pdf_path, similar_lines):
    doc = fitz.open(pdf_path)

    for line in similar_lines:
        for page in doc:
            text_instances = page.search_for(line, hit_max=1)

            for text_instance in text_instances:
                x0, y0, x1, y1 = text_instance
                rect = fitz.Rect(x0, y0, x1, y1)
                underline_annot = page.add_underline_annot(rect)
                underline_annot.colors["stroke"] = (0, 0, 0)  # Установка цвета подчеркивания
                underline_annot.update()

    doc.save(output_path)
    doc.close()

    print("Помеченный PDF-файл создан по пути:", output_path)

# Текст эталона заголовка
header_text = "«Строительство и обустройство скважин куста No 10 Гарюшкинского месторождения»"

# Извлечение полного текста из PDF
doc = fitz.open(pdf_path)
full_text = [line.strip() for page in doc for line in page.get_text().split("\n")]
doc.close()

# Поиск строк с похожим текстом
similar_lines = find_similar_lines(full_text, header_text)

# Проверка орфографии схожих строк
corrected_lines = spell_check_lines(similar_lines)

# Создание помеченного PDF-файла
create_marked_pdf(pdf_path, corrected_lines)
```
